#### Function to get all Data related to Powerplay wrt Bowling. 

```python
def PP_Bowling(i):
    # --------------------->Runs Conceeded<-----------------------------------------------------
    runs_conceeded = i['pplay_s_t2'].sum()
    
    # --------------------->Wickets Lost in PP<-----------------------------------------------
    
    wickets_picked = i['pplay_w_t2'].sum()
    
    
    # ----------------------> Number of Matches <---------------------------------------------
    number_of_matches = len(i)
    
    min_pp_runs_conceeded = i['pplay_s_t2'].min()
    #print("Lowest Runs Conceeded in PP",min_pp_runs_conceeded)
    #print('<--------------------------------------------------------------->')
    max_pp_runs_conceeded = i['pplay_s_t2'].max()
    #print("Highest Runs Conceeded in PP",max_pp_runs_conceeded)
    #print('<--------------------------------------------------------------->')
    max_wickets_picked = i['pplay_w_t2'].max()
    #print(max_wickets_picked)
    
    overs_bowled = len(i)*6
    
    economy = round(runs_conceeded/overs_bowled,2)
    

    Stats_Bowling = pd.DataFrame.from_dict({"Matches Played":number_of_matches,'Powerplay Runs Conceeded':runs_conceeded,'Powerplay Wickets Picked':wickets_picked, "Powerplay Economy":economy, "Lowest Conceeded Runs in Powerplay":min_pp_runs_conceeded,"Highest Conceeded Runs in Powerplay":max_pp_runs_conceeded, "Max. Wickets Picked in Powerplay":max_wickets_picked}, orient='index', columns=['Value'])
    Stats_Bowling = Stats_Bowling.transpose()
    
    #Stats_Bowling=Stats_Bowling.astype('int64')
    Stats_Bowling['Matches Played'] = Stats_Bowling['Matches Played'].astype('int64')
    Stats_Bowling['Powerplay Runs Conceeded'] = Stats_Bowling['Powerplay Runs Conceeded'].astype('int64')
    Stats_Bowling['Powerplay Wickets Picked'] = Stats_Bowling['Powerplay Wickets Picked'].astype('int64')
    #Stats_Bowling['Total Overs Bowled'] = Stats_Bowling['Total Overs Bowled'].astype('int64')
    Stats_Bowling['Lowest Conceeded Runs in Powerplay'] = Stats_Bowling['Lowest Conceeded Runs in Powerplay'].astype('int64')
    Stats_Bowling['Highest Conceeded Runs in Powerplay'] = Stats_Bowling['Highest Conceeded Runs in Powerplay'].astype('int64')
    Stats_Bowling['Max. Wickets Picked in Powerplay'] = Stats_Bowling['Max. Wickets Picked in Powerplay'].astype('int64')
    
    return Stats_Bowling
```
